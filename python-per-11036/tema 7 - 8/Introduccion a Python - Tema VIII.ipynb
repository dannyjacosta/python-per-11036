```python
# Dos diagramas en una misma imagen.
income_2024 = [1274, 1300, 1400, 1045, 1622]
income_2025 = [1300, 1400, 1200, 945, 1722]
etiquetas = ['En', 'Feb', 'Mar', 'Abr', 'May']
x = np.arange(len(etiquetas))

plt.plot(income_2024, label='2024')
plt.plot(income_2025, label='2025')

plt.ylabel('Revenue')
plt.title('Revenues 2024 vs 2025')
plt.xticks(x, etiquetas)
plt.grid(axis='y', linestyle=':')
plt.legend()
plt.show()
```

### Gr√°ficas de Barra

```python
values_x = ['A', 'B', 'C', 'D', 'E']
values_y = np.random.randint(low=1, high=20, size=5)
plt.bar(values_x, values_y, color='skyblue', edgecolor='black')

plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Valores por Categor√≠a')
plt.grid(axis='y', linestyle='--')

for i, value in enumerate(values_y):
    plt.text(i, value + 1, str(value), ha='center')

plt.show()
```

```python
# Dos diagramas en una misma imagen.
income_2024 = [1274, 1300, 1400, 1045, 1622]
income_2025 = [1300, 1400, 1200, 945, 1722]
etiquetas = ['En', 'Feb', 'Mar', 'Abr', 'May']
x = np.arange(len(etiquetas))
width = 0.35

plt.bar(x - width/2, income_2024, width, label='2024', color='skyblue')
plt.bar(x + width/2, income_2025, width, label='2025', color='orange')

plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Revenues 2024 vs 2025')
plt.xticks(x, etiquetas)
plt.grid(axis='y', linestyle=':')
plt.legend()
plt.show()
```

### Varios diagramas en una misma figura

```python
income_2024 = [1274, 1300, 1400, 1045, 1622]
income_2025 = [1300, 1400, 1200, 945, 1722]
etiquetas = ['En', 'Feb', 'Mar', 'Abr', 'May']
x = np.arange(len(etiquetas))
width = 0.35

fig = plt.gcf()
fig.set_size_inches(10,8)

plt.subplot(221)
plt.bar(x - width/2, income_2024, width, label='2024', color='skyblue')
plt.bar(x + width/2, income_2025, width, label='2025', color='orange')
plt.xticks(x, etiquetas)
plt.ylabel('Revenue in thousands')
plt.grid(axis='y', linestyle=':')
plt.legend()

plt.subplot(222)
plt.plot(income_2024, label='2024')
plt.plot(income_2025, label='2025')
plt.xticks(x, etiquetas)
plt.ylabel('Revenue in thousands')
plt.grid(axis='y', linestyle=':')
plt.legend()

plt.subplot(223)
plt.pie(income_2024, labels=etiquetas)

plt.subplot(224)
angles = np.linspace(0, 2 * np.pi, 1000)
tan_values = np.tan(angles)
plt.plot(angles, tan_values)
plt.grid(True)
plt.xlabel('Œ∏ (radians)')
plt.ylabel('tang(Œ∏)')

plt.tight_layout()
plt.show()
```

## Plotly

```python
from plotly.offline import init_notebook_mode
from plotly.offline import iplot
import plotly.graph_objects as go
import numpy as np

init_notebook_mode(connected=True)
```

```python
values = np.array(range(1,100))

data = [go.Scatter(y=values, mode='markers', marker=dict(color='red', symbol='pentagon', size=10))]
iplot(data)
```

```python
fig = go.Figure()
angulos = np.linspace(0, np.pi, 1000)
tangentes = np.tan(angulos)
senos = np.sin(angulos)
fig.add_trace(go.Scatter(x=angulos, y=tangentes, name='tan(x)'))
# fig.add_trace(go.Scatter(x=angulos, y=senos, name='sin(x)'))


fig.update_layout(
    xaxis=dict(
        zeroline=True,
        zerolinewidth=2,
        zerolinecolor='gray',

    ),
    yaxis=dict(
        zeroline=True,
        zerolinewidth=2,
        zerolinecolor='gray',
    ),
    title='Tangent Function',
    xaxis_title='x',
    width=800,
    height=800
)

fig.show()
```

```python
x = np.linspace(0, -10, 100)
y = x ** 2
fig = go.Figure()

fig.add_trace(go.Scatter(x = x, y = y))
fig.update_layout(
    width=600,
    height=900
)
fig.show()
```

```python
fig = go.Figure()
angulos = np.linspace(1, 1000, 200)
tangentes = np.tan(angulos)
senos = np.sin(angulos)
secanntes = 1 / np.cos(angulos)

fig.add_trace(go.Histogram(x=angulos, name='Angulos', xbins=dict(start=0, end=200)))
fig.show()
```

```python
income_2025 = [1300, 1400, 1200, 945, 1722]
etiquetas = ['En', 'Feb', 'Mar', 'Abr', 'May']

fig = go.Figure()
fig.add_trace(go.Bar(x=etiquetas, y=income_2024, name='2024'))
fig.add_trace(go.Bar(x=etiquetas, y=income_2025, name='2025'))
fig.update_layout(width=500, title='Revenue in millions USD')
fig.show()
```
