## Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

### Gr√°ficas de lineas y puntos

```python
valores = np.array([1, 2, 3, 4, 6, 9, 14, 10, 8, 25])
plt.plot(valores, ':r')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
valores_x = np.array([1, 2, 3, 4])
valores_y = np.array([4, 3, 2, 1])
plt.plot(valores_x, valores_y)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
plt.plot(valores_x, valores_y, 'pr')
plt.show()
```

```python
plt.plot(valores_x, valores_y, '--', linewidth=1)
plt.show()
```

```python
values_x1 = np.random.randint(low=1, high=100, size=10)
values_y1 = np.random.randint(low=1, high=100, size=10)
plt.plot(values_x1, values_y1, 'co')

values_x2 = np.random.randint(low=1, high=100, size=10)
values_y2 = np.random.randint(low=1, high=100, size=10)
plt.plot(values_x2, values_y2, ':')
plt.show()
```

```python
# Dos diagramas en una misma imagen.
income_2024 = [1274, 1300, 1400, 1045, 1622]
income_2025 = [1300, 1400, 1200, 945, 1722]
etiquetas = ['En', 'Feb', 'Mar', 'Abr', 'May']
x = np.arange(len(etiquetas))

plt.plot(income_2024, label='2024')
plt.plot(income_2025, label='2025')

plt.ylabel('Revenue')
plt.title('Revenues 2024 vs 2025')
plt.xticks(x, etiquetas)
plt.grid(axis='y', linestyle=':')
plt.legend()
plt.show()
```

### Histogramas

```python
valores = np.random.randn(10000)
plt.hist(valores)
plt.show()
```

```python
#Modificando orientaci√≥n y color
plt.hist(valores, orientation='horizontal', color='green')
plt.show()
```

```python
#Modificando el rango de valores
plt.hist(valores, range=(-1, 1))
plt.show()
```

```python
#Histograma 2D
valores_x = np.random.randint(low=1, high=100, size=1000)
valores_y = np.random.randint(low=1, high=100, size=1000)
#print(valores_x)
#print(valores_y)
plt.hist2d(valores_x, valores_y)
plt.show()
```

### Gr√°ficas de Barra

```python
values_x = ['A', 'B', 'C', 'D', 'E']
values_y = np.random.randint(low=1, high=20, size=5)
plt.bar(values_x, values_y, color='skyblue', edgecolor='black')

plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Valores por Categor√≠a')
plt.grid(axis='y', linestyle='--')

for i, value in enumerate(values_y):
    plt.text(i, value + 1, str(value), ha='center')

plt.show()
```

```python
# Dos diagramas en una misma imagen.
income_2024 = [1274, 1300, 1400, 1045, 1622]
income_2025 = [1300, 1400, 1200, 945, 1722]
etiquetas = ['En', 'Feb', 'Mar', 'Abr', 'May']
x = np.arange(len(etiquetas))
width = 0.35

plt.bar(x - width/2, income_2024, width, label='2024', color='skyblue')
plt.bar(x + width/2, income_2025, width, label='2025', color='orange')

plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Revenues 2024 vs 2025')
plt.xticks(x, etiquetas)
plt.grid(axis='y', linestyle=':')
plt.legend()
plt.show()
```

### Gr√°fico Circular

```python
values = [10, 15, 25,  50]
etiquetas = ['Bug', 'Story', 'Training', 'Support']
plt.pie(values, labels=etiquetas, radius=1)
plt.show()
```

### Varios diagramas en una misma figura

```python
income_2024 = [1274, 1300, 1400, 1045, 1622]
income_2025 = [1300, 1400, 1200, 945, 1722]
etiquetas = ['En', 'Feb', 'Mar', 'Abr', 'May']
x = np.arange(len(etiquetas))
width = 0.35

fig = plt.gcf()
fig.set_size_inches(10,8)

plt.subplot(221)
plt.bar(x - width/2, income_2024, width, label='2024', color='skyblue')
plt.bar(x + width/2, income_2025, width, label='2025', color='orange')
plt.xticks(x, etiquetas)
plt.ylabel('Revenue in thousands')
plt.grid(axis='y', linestyle=':')
plt.legend()

plt.subplot(222)
plt.plot(income_2024, label='2024')
plt.plot(income_2025, label='2025')
plt.xticks(x, etiquetas)
plt.ylabel('Revenue in thousands')
plt.grid(axis='y', linestyle=':')
plt.legend()

plt.subplot(223)
plt.pie(income_2024, labels=etiquetas)

plt.tight_layout()
plt.show()
```
